```python
import pickle
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
boolean_cat = df[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']]
boolean_cat.head()
```

```python
def binary_map(x):
    return x.map({'yes': 1, "no": 0})
boolean_cat = boolean_cat.apply(binary_map)
boolean_cat.head()
```

```python
category = df[['furnishingstatus']]
category.head()
```

```python
def cat_map(x):
    return x.map({'furnished': 1, "semi-furnished": 2, "unfurnished": 3})
category = category.apply(cat_map)
category.head()
```

```python
value = df[['area','bedrooms','bathrooms','stories','parking']]
Y = df[['price']]
X = pd.concat([boolean_cat,category,value],axis=1)
X.head()
```

```python
x = X.values
y = Y.values.reshape(-1)
```

```python
y
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x,y)
```

```python
with open("homework/model.pkl", "wb") as f:
    pickle.dump(model, f)
```
